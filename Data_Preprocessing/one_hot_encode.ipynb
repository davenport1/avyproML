```python
# Import necessary packages.
import pandas as pd
```

```python
# Our total span, across multiple seasons.
years = ("17", "18", "19")
```

```python
# Header names representing the one-hot encoding of danger ratings and problem type.
header_names = ["date", 
                "normal_caution", "wind_slab", 
                "persistent_slab", 
                "loose_wet", "deep_slab", "storm_slab",
                "atl_LOW", "atl_MOD", "atl_CONS", "atl_HIGH",
                "ntl_LOW", "ntl_MOD", "ntl_CONS", "ntl_HIGH",
                "btl_LOW", "btl_MOD", "btl_CONS", "btl_HIGH", ]

df = pd.DataFrame(columns = header_names)
```

```python
for year in years:
    title = "cleaned_avy_" + year + ".csv"
    scraped = pd.read_csv(title)

    for index, row in scraped.iterrows():
        date = row["date"]
        entry = [date]

        # 100000 == normal_caution
        # 010000 == wind_slab
        # 001000 == persistent_slab
        # 000100 == loose_wet
        # 000010 == deep_slab
        # 000001 == storm_slab
        if row["problem_1"] == " Normal Caution":
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
        elif row["problem_1"] == " Wind Slab":
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
        elif row["problem_1"] == " Persistent Slab":
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)
        elif row["problem_1"] == " Loose Wet":
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
        elif row["problem_1"] == " Deep Slab":
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
        elif row["problem_1"] == " Storm Slab":
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
        
        # 1000 == low
        # 0100 == moderate
        # 0010 == considerate
        # 0001 == high
        # SEEMS INEFFICIENT?!?
        # For above treeline.
        if row["above_treeline"] == 4:
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
        elif row["above_treeline"] == 3:
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
        elif row["above_treeline"] == 2:
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
        else:
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)

        # For near treeline.
        if row["near_treeline"] == 4:
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
        elif row["near_treeline"] == 3:
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
        elif row["near_treeline"] == 2:
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
        else:
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)

        # For below treeline.
        if row["below_treeline"] == 4:
            entry.append(0)
            entry.append(0)
            entry.append(0)
            entry.append(1)
        elif row["below_treeline"] == 3:
            entry.append(0)
            entry.append(0)
            entry.append(1)
            entry.append(0)
        elif row["below_treeline"] == 2:
            entry.append(0)
            entry.append(1)
            entry.append(0)
            entry.append(0)
        else:
            entry.append(1)
            entry.append(0)
            entry.append(0)
            entry.append(0)
        
        df.loc[len(df)] = entry
```

#### Create Output CSV

```python
# Sort by date.
df['date'] = pd.to_datetime(df.date, infer_datetime_format = True)
df.sort_values(by = 'date', ascending = True, inplace = True)

# Turn DataFrame into CSV.
df.to_csv("cleaned_avy.csv", index=False)
```
