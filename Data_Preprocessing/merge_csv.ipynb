#### Imports

```python
# Import necessary packages.
import pandas as pd
```

#### Format Scraped Data

```python
# Three seasons worth of overall avy risk ratings, as well as problem types.
df1 = pd.read_csv("cleaned_avy.csv")

# Combine three seasons worth into one DataFrame.
first_q = ["date", 
           "normal_caution", "wind_slab", 
           "persistent_slab", "loose_wet", 
           "deep_slab", "storm_slab",
           "atl_LOW", "atl_MOD", "atl_CONS", "atl_HIGH",
           "ntl_LOW", "ntl_MOD", "ntl_CONS", "ntl_HIGH",
           "btl_LOW", "btl_MOD", "btl_CONS", "btl_HIGH" ]

first = pd.DataFrame()
for header in first_q:
    col = df1[header]
    first[header] = col

first.head()
```

#### Format Aspect Data

```python
df2 = pd.read_csv("aspects.csv")
second_q = ["date", "N_btl", "N_ntl", "N_atl",
            "NE_btl", "NE_ntl", "NE_atl",
            "E_btl", "E_ntl", "E_atl",
            "SE_btl", "SE_ntl", "SE_atl",
            "S_btl", "S_ntl", "S_atl",
            "SW_btl", "SW_ntl", "SW_atl",
            "W_btl", "W_ntl", "W_atl",
            "NW_btl", "NW_ntl", "NW_atl"]

# Add headers.
second = pd.DataFrame()
for header in second_q:
    col = df2[header]
    second[header] = col

second.head()
```

#### Format Size Data

```python
df3 = pd.read_csv("sizes.csv")
third_q = ["date", "small", "large", "very_large", "historic"]

# Add headers.
third = pd.DataFrame()
for header in third_q:
    col = df3[header]
    third[header] = col

third.head()
```

#### Format Likelihood Data

```python
df4 = pd.read_csv("likelihoods.csv")
fourth_q = ["date", "unlikely", "possible", "likely", "very_likely", "certain"]

# Add headers.
fourth = pd.DataFrame()
for header in fourth_q:
    col = df4[header]
    fourth[header] = col

fourth.head()
```

#### Merge DataFrames

```python
# Merge our four dataframes by date.
first_sec = pd.merge(first, second, on="date")
first_sec_third = pd.merge(first_sec, third, on = "date")
data = pd.merge(first_sec_third, fourth, on = "date")
data.head()
```

#### Create Y-train Output

```python
data.to_csv("Y_train.csv", index=False)
```
